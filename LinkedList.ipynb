# Linked List Implementation in Python

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None
        
    # Insert element at start of list
    def insertAtStart(self, data):
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node

    # Insert element at end of list
    def insertAtEnd(self, data):
        if self.findLength() == 0 :
            self.insertAtStart(data)
            return
        new_node = Node(data)
        temp = self.head
        while temp.next != None :
            temp = temp.next
        temp.next = new_node

    # calculate length of list
    def findLength(self):
        count = 0
        temp = self.head
        while temp:
            count += 1
            temp = temp.next
        return count
    
    # insert at a specific location in list
    def insertAtLocation(self, location, data):
        if location not in range(1, self.findLength() + 1): # 1 is added because range exclude last value
            print( location, " is not a location in linked list")
            return None
        elif location == 1 :
            print("in else condition")
            self.insertAtStart(data)
            return None
        new_node = Node(data)
        temp = self.head
        current_location = 1
        while current_location < location -1 :
            current_location += 1
            temp = temp.next

        new_node.next = temp.next
        temp.next = new_node

    # delete from start
    def deleteFromStart(self):
        # temp = self.head => Python garbage collector will automatically delete the memory of Nodes without any reference
        self.head = self.head.next
        # del temp

    # delete from last
    def deleteFromEnd(self):
        current_node = self.head
        previous_node = None
        while current_node.next != None :
            previous_node = current_node
            current_node = current_node.next

        previous_node.next = None 

    # delete from specific location in list
    def deleteFromLocation(self, location):
        if location not in range(1, self.findLength() + 1): # 1 is added because range exclude last value
            print( location, " is not a location in linked list")
            return None
        elif location == 1 :
            self.deleteFromStart()
            return None

        current_node = self.head
        current_location = 1
        while current_location < location - 1 :
            current_location += 1
            current_node = current_node.next

        current_node.next = current_node.next.next
        
    # display list items
    def display(self):
        temp = self.head
        while temp != None :
            print(temp.data, end=" ")
            temp = temp.next



# testing functionality

# insert at start
list = LinkedList()
# list.insertAtStart(3)
# list.insertAtStart(2)
# list.insertAtStart(1)

# insert at end
list.insertAtEnd(1)
list.insertAtEnd(2)
list.insertAtEnd(3)
list.insertAtEnd(4)
list.insertAtEnd(5)

# list.insertAtLocation(3,21) # first arg is location second is data

# delete from start
# list.deleteFromStart()
# list.deleteFromStart()

# delete from end
# list.deleteFromEnd()
# list.deleteFromEnd()

# delete from specific location
# list.deleteFromLocation(6) # pass location arg

print("Length of list = ", list.findLength())
list.display()
```
